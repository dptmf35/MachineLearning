```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(c_vec_cat, c_vec_cat).argsort()[:,::-1]
```

```python
print(cos_sim)
```

```python
cos_sim.shape
```

```python
def get_recommend_media(df, title, top=5) :
    target_media_index = df[df['제목']==title].index.values
    
    simm_index = cos_sim[target_media_index,:top].reshape(-1)
    
    simm_index = simm_index[simm_index != target_media_index]
    
    result = df.iloc[simm_index].sort_values('조회수',ascending = False)[:10]
    
    return result
```

```python
from surprise.model_selection import KFold

options = {
    'method' : 'als',
    'n_epochs' : 5,
    'reg_u' : 12,
    'reg_i' : 5
}

algo = surprise.BaselineOnly(options)

np.random.seed(0)
acc = np.zeros(3)
cv = KFold(3)

for i, (trainset, testset) in enumerate(cv.split(data)) : 
    algo.fit(trainset)
    predictions = algo.test(testset)
    acc[i] = surprise.accuracy.rmse(predictions, verbose=True)

acc.mean()
```

```python
from surprise.model_selection import cross_validate

cross_validate(algo,data)
```

```python
#COSINE

sim_options = {'name' : 'cosine'}
algo = surprise.KNNBasic(sim_options=sim_options)
cross_validate(algo,data)['test_mae'].mean()
```

```python
#PEARSON

sim_options = {'name' : 'pearson'}
algo = surprise.KNNBasic(sim_options=sim_options)
cross_validate(algo,data)['test_mae'].mean()
```

```python
#PEARSON_BASELINE

sim_options = {'name': 'pearson_baseline'}
algo = surprise.KNNBasic(sim_options=sim_options)
cross_validate(algo, data)["test_mae"].mean()
```

```python
#KNN Bbaseline

sim_options = {'name': 'pearson_baseline'}
algo = surprise.KNNWithMeans(sim_options=sim_options)
cross_validate(algo, data)["test_mae"].mean()
```

```python
#SVD

%%time
algo = surprise.SVD(n_factors=100)
cross_validate(algo, data)["test_mae"].mean()
```

```python
#NMF

%%time
algo = surprise.NMF(n_factors=100)
cross_validate(algo, data)["test_mae"].mean()
```
