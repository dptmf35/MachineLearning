```python
import numpy as np
```

### 독립변수 x 생성

```python
x = np.random.uniform(low = 0.0, high = 1.0, size = 100) # 0~~1사이 100개의 랜덤 값 생성
print(x.shape)
x[:3] #데이터 3개만 출력
```

### 종속변수 y = wx + b 생성

```python
'''w = 10, b = 2'''
y = 10*x + 2
print(y.shape)
y[:3]
```

```python
w = np.random.uniform(low=0.0, high=1.0)
b = np.random.uniform(low=0.0, high=1.0)

learning_rate = 3.083
num_epoch = 300

w,b
```

```python
for epoch in range(num_epoch) : ##몇번학습??
    print(f"{epoch:2}, w = {w:.4f}, b = {b:.4f}")
    
    """랜덤설정된 w,b값의 결과"""
    y_pre = w * x + b
    
    """MSE(평균제곱오차)"""
    mse = ((y_pre - y)**2).mean()
    print(mse)
    
    error = np.abs(y_pre - y).mean()
    if error < 0.001 :
        break
       
    ##(wx+b - y)**2 >> 2(wx+b-y)x : 미분
    
    w = w - learning_rate*((y_pre - y)*x).mean()
    b = b - (y_pre - y).mean()
    ## w = w - ((a * dw/d) * cost(w))    : cost(w) == (wx+b-y)**2
 
print("-"*50)
print(f"{epoch:2}, w = {w:.4f}, b = {b:.4f}")
```

### 독립 변수 2개 생성

```python
x1 = np.random.uniform(low= 0.0 , high = 1.0, size = 100)
x2 = np.random.uniform(low= 0.0 , high = 1.0, size = 100)

y2 = 3*x1 + 7*x2 + 10  ## w1 = 3, w2 = 7, b = 10
print(y2.shape)
print(x1[:2])
print(x2[:2])
print(y2[:2])
```

```python
w1 = np.random.uniform(low= 0.0 , high = 1.0)
w2 = np.random.uniform(low= 0.0 , high = 1.0)
b = np.random.uniform(low= 0.0 , high = 1.0)
## learning rate와 epoch 설정 해주기
learning_rate = 1.33
num_epoch = 200

w1, w2, b
```

```python
for epoch in range(num_epoch) :
    print(f"{epoch:2}, w1 = {w1:.4f}, w2 = {w2:.4f}, b = {b:.4f}")
    
    y2_pre = w1*x1 + w2*x2 + b
    mse = ((y2_pre - y2)**2).mean()
    print(mse)
    
    error = np.abs(y2_pre - y2).mean()
    
    if error < 0.001 :
        break
    
    w1 = w1 - learning_rate*((y2_pre - y2)*x1).mean()
    w2 = w2 - learning_rate*((y2_pre - y2)*x2).mean()
    b = b - (y2_pre - y2).mean()
print('-'*50)
print(f"{epoch:2}, w1 = {w1:.4f}, w2 = {w2:.4f}, b = {b:.4f}")    
```

```python
import pandas as pd
```

```python
x_train = pd.DataFrame([x1,x2], index=["x1", "x2"])
x_train = x_train.T
x_train.head()
```

```python
y_train = pd.Series(y2)
print(y_train.shape)
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression ##선형회귀모델
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```
