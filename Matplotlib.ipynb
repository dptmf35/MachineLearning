```python
data3.head()
```

```python
# na를 앞 값으로 채움
df = data3.fillna(method = 'ffill')
```

```python
# 서울에서 다른 지역으로 이동 데이터만 추출하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별']!= '서울특별시')
# 전출지가 서울특별시이고 전입지가 서울특별시가 아닌 행 추출(불리언인덱싱)
df_seoul = df[mask]
```

```python
df_seoul.head()
# 전출지는 모두 서울특별시기 때문에 의미없는 데이터 -- > 전출지별 컬럼 삭제
df_seoul = df_seoul.drop(['전출지별'], axis = 1)
```

```python
df_seoul.rename({'전입지별' : '전입지'}, axis = 1, inplace = True) #컬럼명 다시 설정 (전입지별 -- > 전입지)
df_seoul.set_index('전입지', inplace= True) #전입지로 index 설정
```

```python
df_seoul.head()
```

```python
df_seoul.index
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
sr_one = df_seoul.loc['경기도']
sr_one # < -- 딕셔너리 타입 데이터
```

```python
plt.plot(sr_one.index , sr_one.values) # x축을 sr_one의 인덱스(연도)로 , y축을 sr_one의 값(인구수)으로 설정
```

```python
plt.xlabel('기간')
plt.ylabel('이동한 인구수')
plt.title('서울에서 경기도로 이동한 인구 수')
plt.plot(sr_one.index , sr_one.values)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
sns.set_style('darkgrid')
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.regplot( x = 'age', # x 축 변수
            y = 'fare', # y 축 변수
            data = titanic, # dataset 
            ax = ax1 #axe 객체 
           )
sns.regplot( x = 'age', # x 축 변수
            y = 'fare', # y 축 변수
            data = titanic, # dataset 
            ax = ax2, #axe 객체 
            fit_reg = False # 회귀선 미표시
           )
plt.show()
```

```python

sns.set_style('darkgrid')
fig = plt.figure(figsize = (20,5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.distplot(titanic['fare'], ax = ax1)
sns.distplot(titanic['fare'], ax = ax2, hist = False)
sns.distplot(titanic['fare'], ax = ax3, kde = False)

```

```python
table = titanic.pivot_table(index = ['sex'], columns= ['class'], aggfunc='size')
sns.heatmap(table,
            annot = True, fmt = 'd',
           cmap = 'YlGnBu',
           linewidth = 5,
           cbar = False)
plt.show()
```

```python
sns.set_style( 'whitegrid')
fig = plt.figure( figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.stripplot( x = 'class' , 
             y = 'age',
              data = titanic,
              ax = ax1
             )
sns.stripplot( x = 'class' , 
             y = 'age',
              data = titanic,
              ax = ax2
             )
```

## 위치 좌표 데이터를 읽어와서 지도에 표시하기

```python
df = pd.read_excel("서울지역 대학교 위치.xlsx")
univ = folium.Map(location = [37.55 , 126.98], zoom_start = 12)
```

```python
df.head()
```

```python
for name, lat, lng in zip(df.대학명, df.위도, df.경도) :
    folium.Marker([lat, lng] , popup = name ).add_to(univ)
##마커로 위도,경도 따라서 지도에 표기
```

```python
univ.save('univ_map.html')
```

```python
import json
```

```python
file_path = '경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col = '구분')
df.columns = df.columns.map(str)

geo_path = '경기도행정구역경계.json'


try :
    geo_data = json.load(open(geo_path, encoding = 'utf-8'))
except :
    geo_data = json.load(open(geo_path, encoding = 'utf-8-sig'))
    
#경기도 지도 만들기
g_map = folium.Map( location = [37.5502, 126.982],
                  tiles= 'Stamen Terrain', zoom_start =9)
year = '2007'

folium.Choropleth( geo_data = geo_data,
                 data = df[year],
                 columns = [df.index, df[year]],
                 fill_color = 'YlOrRd', fill_opacitiy = 0.7, line_opacity= 0.3,
                 threadhold_scale = [10000,100000,300000,500000,7000000], 
                 key_on = 'feature.properties.name').add_to(g_map)
```
