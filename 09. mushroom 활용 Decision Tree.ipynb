## 랜덤포레스트

```python
mush = pd.read_csv("./data/mushroom.csv")

x = mush.iloc[:,1:]
y = mush.iloc[:,0]
```

```python
x = pd.get_dummies(x, prefix = x.columns)
dic = {'p':0,'e':1}
y =y.map(dic)
```

```python
x_train, x_test, y_train, y_test = \
train_test_split(x, y, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=30) ##n_estimator = 30~100 
model.fit(x_train, y_train)


print("훈련 세트 정확도: {:.3f}".format(model.score(x_train, y_train)))
print("테스트 세트 정확도: {:.3f}".format(model.score(x_test, y_test)))
```

```python
featlist = list(zip(model.feature_importances_, x.columns))
featlist.sort(reverse = True)
feat = featlist[:15]
feat_df = pd.DataFrame(feat, columns = ['score', 'feature'])
plt.barh(feat_df['feature'], feat_df['score'])
```
